# Sleep health

```python
import pandas as pd

df1 = pd.read_csv("./data/employee_survey.csv")
df2 = pd.read_csv("./data/job_stress_merge.csv")
```

```python
df1 = df1.rename(columns={"Dept": "occupation"})
df2 = df2.rename(columns={"occupation1": "occupation"})

df2_subset = df2[[
    "occupation",
    "work_schedule_variability",
    "work_schedule_variability_logit",
    "pause_work",
    "pause_work_logit",
    "work_pace",
    "work_pace_pca"
]]

df_merged = df1.merge(
    df2_subset,
    on="occupation",
    how="left"
)
df_merged.head()
```

```python
df_merged.to_csv("./data/employee_survey_merge.csv", index=False)
```
